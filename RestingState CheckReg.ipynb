```python
templatedir = '/Users/cplanting/Projects/RestingState/template/'
subdir = '/Users/cplanting/Projects/RestingState/n3806/'
mni_img = nib.load(op.join(templatedir,"MNI152_T1_1mm_brain.nii.gz"))
#mni_img = nib.load(op.join(templatedir,"avg152T1_brain.nii.gz"))
mni_data, mni_aff = mni_img.get_data().astype(float), mni_img.get_affine()
sef_img = nib.load(op.join(subdir,"highres2template.nii.gz"))
sef_data, sef_aff = sef_img.get_data(), sef_img.get_affine()
sef_data[sef_data < 1] = 0
```
