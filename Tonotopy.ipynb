## Tonotopy

```python
import scipy.io
import os.path as op
import numpy as np
import nibabel
import scipy.io
import os.path as op
import numpy as np
import matplotlib.pyplot as plt
from mayavi import mlab
from nipy.labs.viz import plot_map, mni_sform, coord_transform
import nibabel
import os

# a scale function to normalize a matrix
def scale(matrix):
    from numpy import mean, std
    return (matrix - mean(matrix, axis=0)) / std(matrix, axis=0)
```

```python
resdir = '/Users/crislanting/Dropbox/tamac_share/confiles/freq'
statfile = 'spmT_0001.img'
filelist = ['con_0003.img','con_0004.img','con_0005.img','con_0006.img']

#3d cut-out
x,y,z = (-40, -23, 5)
z_range = (-6,-3,0,3,6,9)
thresh = 5
fs = 12

img = nibabel.load(op.join(resdir,statfile))
data = img.get_data()
affine = img.get_affine()
data_thresh = np.ma.masked_less(data,thresh)
mask = data > thresh
nvox = np.sum(np.sum(np.sum(mask,axis=-1),axis=-1),axis=-1)
plot_map(data_thresh,affine,threshold=thresh,figure=2,cut_coords=(x,y,z),cmap=plt.cm.autumn)

tmp = np.zeros((nvox,len(filelist)))
n = 0
for i in filelist:
	img = nibabel.load(op.join(resdir,i))
	img_data = img.get_data()
	tmp[:,n] = img_data[mask]
	n = n +1
    
#how many voxels and conditions?
print tmp.shape
```

```python
# rank-order per column (maximum: rank 1; minimum: 0)
order = tmp.argsort(axis=0)
rankorder = order.argsort(axis=0)
tmp_sort = rankorder/(np.float(tmp.shape[0])-1)

# assign maximum rank across frequency as best-frequency
rank = np.argmax(tmp_sort,axis = -1)

# interpolation using betas as weights:
# calculate best frequency w.r.t. 1000 Hz -> tmp_f goes from [-2 +3] or [250Hz 8000Hz]; see Eq 1. of Langers (2013), 
# "Assessment of tonotopically organised subdivisions in human auditory cortex using volumetric and surface-based cortical alignments."
tmp_f = (-2*tmp[:,0]**2 -0.5*tmp[:,1]**2 + 1.5*tmp[:,2]**2 + 3*tmp[:,3]**2)/(np.sum(tmp**2,axis=-1))

#plot distribution of interpolated frequencies
figure1 = plt.figure(1,figsize=[10,5])
ax1 = figure1.add_subplot(121, autoscale_on=False, xlim=[-2,3], ylim=[0,1600])
plt.hist(tmp_f,30)
ax1.set_xlabel('Frequency (Hz)',fontsize=fs)
ax1.set_ylabel('Frequency',fontsize=fs)
xlocs = [-2,-1,0,1,2,3]
xlabels = ['250','500','1000','2000','4000','8000']
xticks(xlocs, xlabels,fontsize=fs)

ylocs = [0,400,800,1200,1600]
yticks(ylocs, ylocs,fontsize=fs)

#plot distribution of ranks
ax2 = figure1.add_subplot(122, autoscale_on=True)
plt.hist(rank,4)
t = np.array([0,0.75,1.5,2.25])
xlocs = t+(0.75/2.0)
xlabels = ['250','750','3000','8000']
xticks(xlocs, xlabels,fontsize=fs)
ax2.set_xlabel('Rank',fontsize=fs)
ax2.set_ylabel('Frequency',fontsize=fs)
ylocs = [0,4000]
yticks(ylocs, ylocs,fontsize=fs)
figure1.tight_layout();
```

```python
#write rank-ordered data to file
img_rank = np.zeros(img_data.shape)

#rank goes from 0-3; introduce offset of 1 for viewing
img_rank[mask] = 1+rank[:]
fname_out = op.join(resdir,'rank_ordered.nii.gz')
img_out = nibabel.Nifti1Image(img_rank, affine)
nibabel.save(img_out, fname_out)

#smooth with fslmaths
filein = op.join(resdir,'rank_ordered.nii.gz')
fileout = op.join(resdir,'rank_ordered_smooth_15.nii.gz')
sigma = '1.5 '
comm = "fslmaths " + filein + " -s " + sigma + fileout
os.system(comm)
```

```python
#write frequency-interpolation-data to file
#best frequency w.r.t. 1000 Hz goes from [-2 +3]; introduce an offset of 2 for viewing
img_f = np.zeros(img_data.shape)
img_f[mask] = 2 + tmp_f[:]
fname_out = op.join(resdir,'f.nii.gz')
img_out = nibabel.Nifti1Image(img_f, affine)
nibabel.save(img_out, fname_out)

#smooth with fslmaths
filein = op.join(resdir,'f.nii.gz')
fileout = op.join(resdir,'f_smooth_15.nii.gz')
sigma = '1.5 '
comm = "fslmaths " + filein + " -s " + sigma + fileout
os.system(comm)
```

```python
# perform SVD decomposition on beta-weights: X= U Σ V
# where: tmp_scaled = m (voxels) × n (conditions)
tmp_scaled = scale(tmp)
U, S, V = np.linalg.svd(tmp_scaled)
V = V.T

f, (ax1, ax2) = subplots(ncols=2)
ax1.plot(S, 'ok-',lw=2)
ax1.set_xlabel('Column')
ax1.set_ylabel('Singular value')
ax2.plot(100* S**2 / np.sum(S**2), 'ok-',lw=2)
ax2.set_xlabel('Column')
ax2.set_ylabel('Percent of variance explained')

f.tight_layout();
```

```python
#AM
resdir = '/Users/crislanting/Dropbox/tamac_share/confiles/AM/'
filelist = ['con_0003.img','con_0004.img','con_0005.img','con_0006.img','con_0007.img']

tmpAM = np.zeros((nvox,len(filelist)))
n = 0
for i in filelist:
	img = nibabel.load(op.join(resdir,i))
	img_data = img.get_data()
	tmpAM[:,n] = img_data[mask]
	n = n +1
    
#how many voxels?
print tmpAM.shape

tmp_AM_inter = (1*tmpAM[:,1]**2 + 2*tmpAM[:,2]**2 + 3*tmpAM[:,3]**2 + 4*tmpAM[:,4]**2)/(np.sum(tmpAM[:,1:]**2,axis=-1))
#categories rather than rates. 

#plot distribution of interpolated frequencies
figure4 = plt.figure(4,figsize=[5,5])
ax1 = figure4.add_subplot(111, autoscale_on=False, xlim=[2,3], ylim=[0,3000])
plt.hist(tmp_AM_inter,30)
ax1.set_xlabel('AM rate (Hz)',fontsize=fs)
ax1.set_ylabel('Frequency',fontsize=fs)

img_AM = np.zeros(img_data.shape)
img_AM[mask] = tmp_AM_inter
fname_out = op.join(resdir,'AM.nii.gz')
img_out = nibabel.Nifti1Image(img_AM, affine)
nibabel.save(img_out, fname_out)
```
