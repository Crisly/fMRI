```python
import scipy.io
import os.path as op
import numpy as np
import nibabel
import scipy.io
import os.path as op
import numpy as np
import matplotlib.pyplot as plt
from mayavi import mlab
from nipy.labs.viz import plot_map, mni_sform, coord_transform
import nibabel
import os

# a scale function to normalize a matrix
def scale(matrix):
    from numpy import mean, std
    return (matrix - mean(matrix, axis=0)) / std(matrix, axis=0)
```

```python
resdir = '/Users/crislanting/Dropbox/tamac_share/confiles/freq'
statfile = 'spmT_0001.img'
filelist = ['con_0003.img','con_0004.img','con_0005.img','con_0006.img']

#3d cut-out
x,y,z = (-40, -23, 5)
z_range = (-6,-3,0,3,6,9)
thresh = 5.45
fs = 12

img = nibabel.load(op.join(resdir,statfile))
data = img.get_data()
affine = img.get_affine()
data_thresh = np.ma.masked_less(data,thresh)
mask = data > thresh
nvox = np.sum(np.sum(np.sum(mask,axis=-1),axis=-1),axis=-1)

#store all beta-values in array 'beta'; beta[:,0] contains beta's of con_0003.img etc.
beta = np.zeros((nvox,len(filelist)))
n = 0
for i in filelist:
	img = nibabel.load(op.join(resdir,i))
	img_data = img.get_data()
	beta[:,n] = img_data[mask]
	n = n +1    
#how many voxels and conditions?
print beta.shape

#vmin=5,vmax=12,
#facecolor='k', edgecolor='k'
figure2 = plt.figure(2,figsize=[8,4])
slicer = plot_map(data_thresh,affine,threshold=thresh,figure=2,cut_coords=(x,y,z),vmin=5,vmax=15.5,cmap=plt.cm.hot)
img = plt.gca().get_images()[1] # Or [0] if you have no anat image
cax = plt.axes([1.05, 0.1, .04, .8])
cbar = plt.colorbar(img, cax=cax, ticks=[5.5,10,15])
cbar_tick_obj = plt.getp(cbar.ax.axes, 'yticklabels')                
plt.setp(cbar_tick_obj, color='k')
cbar.set_clim(5, 16.0)
cbar.set_label(r't-value',color='k')
#cbar.set_ticks(['t = 5.5','t = 10', 't =15'])
plt.savefig(op.join(resdir, 't-value.png'), format='png')
```

```python
# rank-order per column (maximum: rank 1; minimum: 0)
order = beta.argsort(axis=0)
rankorder = order.argsort(axis=0) #most 'active' voxel gets highest rank
tmp_sort = rankorder/(np.float(beta.shape[0])-1)
# assign maximum rank across frequency as best-frequency
rank = np.argmax(tmp_sort,axis = -1)

# interpolation using betas as weights:
# calculate best frequency w.r.t. 1000 Hz -> tmp_f goes from [-2 +3] or [250Hz 8000Hz]; see Eq 1. of Langers (2013), 
# "Assessment of tonotopically organised subdivisions in human auditory cortex using volumetric and surface-based cortical alignments."
tmp_f = (-2*beta[:,0]**2 -0.5*beta[:,1]**2 + 1.5*beta[:,2]**2 + 3*beta[:,3]**2)/(np.sum(beta**2,axis=-1))

#plot distribution of interpolated frequencies
figure1 = plt.figure(1,figsize=[10,5])
ax1 = figure1.add_subplot(121, autoscale_on=False, xlim=[-2,3], ylim=[0,1600])
plt.hist(tmp_f,30)
ax1.set_xlabel('Frequency (Hz)',fontsize=fs)
ax1.set_ylabel('Frequency',fontsize=fs)

xlocs = [-2,-1,0,1,2,3]
xlabels = ['250','500','1000','2000','4000','8000']
xticks(xlocs, xlabels,fontsize=fs)
ylocs = [0,400,800,1200,1600]
yticks(ylocs, ylocs,fontsize=fs)

#plot distribution of ranks
ax2 = figure1.add_subplot(122, autoscale_on=True)
plt.hist(rank,4)
t = np.array([0,0.75,1.5,2.25])
xlocs = t+(0.75/2.0)
xlabels = ['250','750','3000','8000']
xticks(xlocs, xlabels,fontsize=fs)
ax2.set_xlabel('Rank',fontsize=fs)
ax2.set_ylabel('Frequency',fontsize=fs)
ylocs = [0,4000]
yticks(ylocs, ylocs,fontsize=fs)
figure1.tight_layout(); 
```

```python
# perform SVD decomposition on beta-weights: X = U Σ V
# where: tmp_scaled = m (voxels) × n (conditions)
beta_scaled = scale(beta)
U, S, V = np.linalg.svd(beta_scaled)
V = V.T

f, (ax1, ax2) = subplots(ncols=2)
ax1.plot(S, 'ok-',lw=2)
ax1.set_xlabel('Column')
ax1.set_ylabel('Singular value')
ax2.plot(100* S**2 / np.sum(S**2), 'ok-',lw=2)
ax2.set_xlabel('Column')
ax2.set_ylabel('Percent of variance explained')
f.tight_layout();
```

```python
# make linear pattern and determine best fit. parameter w determines weight for each of the PCs
ym = np.array([-1,-1/3.,1/3.,1])
xr = np.array([0,1,2,3])
model = np.array([V[:,1], V[:,2],V[:,3]])
w = np.linalg.lstsq(model.T,ym)[0] # obtaining the parameters
xr = np.array([0,1,2,3]) + 0.1

figure2 = plt.figure(2,figsize=[7,7])
ax1 = figure2.add_subplot(221, autoscale_on=False, xlim=[0,4], ylim=[-1.4,1.4])
plt.bar(xr,V[:,0],label='R1')
labelfit = "{:.2f}".format(w[0])  + '* R2 + ' + "{:.2f}".format(w[1]) + '* R3 + ' + "{:.2f}".format(w[2])  + '* R4'
plt.bar(xr+0.1,w[0]*V[:,1]+w[1]*V[:,2]+w[2]*V[:,3],color='r',label=labelfit)
legend(loc='best')
ax1.set_xlabel('Frequency',fontsize=fs)

xlocs = [0.5,1.5,2.5,3.5]
xlabel = ['250','750','3000','8000']
xticks(xlocs, xlabels,fontsize=fs)
ax1.set_xlabel('Frequency',fontsize=fs)
ax1.set_title('R1')

ax2 = figure2.add_subplot(222, autoscale_on=False, xlim=[0,4], ylim=[-1,1])
plt.bar(xr,V[:,1])
xticks(xlocs, xlabels,fontsize=fs)
ax2.set_xlabel('Frequency',fontsize=fs)
ax2.set_title('R2')

ax3 = figure2.add_subplot(223, autoscale_on=False, xlim=[0,4], ylim=[-1,1])
plt.bar(xr,V[:,2])
xticks(xlocs, xlabels,fontsize=fs)
ax3.set_xlabel('Frequency',fontsize=fs)
ax3.set_title('R3')

ax4 = figure2.add_subplot(224, autoscale_on=False, xlim=[0,4], ylim=[-1,1])
plt.bar(xr,V[:,3])
xticks(xlocs, xlabels,fontsize=fs)
ax4.set_xlabel('Frequency',fontsize=fs)
ax4.set_title('R4')

figure2.tight_layout();
```

```python
from sklearn.neighbors import BallTree
from matplotlib.tri import Triangulation, UniformTriRefiner

def NN(X,loc):
	ball_tree = BallTree(X, leaf_size=2)  
	dist, ind = ball_tree.query(X[loc], k=5)
	return ind
	
def calc_gradient(nn_idx):
	vals = values[nn_idx]
	xx = x[nn_idx]
	yy = y[nn_idx]
	diff = (vals - vals[0][0]).flatten()
	diff_x = (xx - xx[0][0]).flatten()
	diff_y = (yy - yy[0][0]).flatten()
	model = np.array([diff_x,diff_y])
	gradx,grady = linalg.lstsq(model.T,diff)[0]
	return gradx,grady

x = np.random.uniform(0,100,1000)
y = np.random.uniform(0,100,1000)
values = np.sqrt(x**2 + y**2)

coords = np.array([x,y]).T	
angle = np.zeros(len(values))
for i in range(len(x)):
	nn_idx = NN(coords,i)
	gradx,grady = calc_gradient(nn_idx)
	da = 180*np.arctan(grady/gradx)/np.pi
	angle[i] = da 

triang = Triangulation(x, y)
```
